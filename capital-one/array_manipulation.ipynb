```python
# Given an array a, your task is to output an array b of the same length by applying the following transformation:

# For each i from 0 to a.length - 1 inclusive, b[i] = a[i - 1] + a[i] + a[i + 1]
# If an element in the sum a[i - 1] + a[i] + a[i + 1] does not exist, use 0 in its place
# For instance, b[0] = 0 + a[0] + a[1]
```

```python
class Solution:

    def __init__(self, nums):
        self.nums = nums

    # b[i] = a[i - 1] + a[i] + a[i + 1]
    def solve(self):
        result = []
        if not self.nums:
            return result
        curr = self.nums[0] + self.nums[1] if len(self.nums) > 1 else 0
        result.append(curr)
        for i in range(1, len(self.nums) - 1):
            curr = curr + self.nums[i+1]
            result.append(curr)
            curr = curr - self.nums[i-1]
        result.append(curr)
        return result

execute = Solution([4, 0, 1, -2, 3])
answer = execute.solve()
print(answer)
```
