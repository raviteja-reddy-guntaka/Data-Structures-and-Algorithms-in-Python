### Question 2: String Pattern Matching - Anagram Groups

Given an array of strings strs, group the anagrams together. You can return the answer in any order.

An anagram is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

Example: For strs = ["eat", "tea", "tan", "ate", "nat", "bat"]:

The output should be [["eat", "tea", "ate"], ["tan", "nat"], ["bat"]].

Explanation:

"eat", "tea", and "ate" are anagrams (all contain 'e', 'a', 't')
"tan" and "nat" are anagrams (all contain 't', 'a', 'n')
"bat" stands alone
This problem tests your understanding of string manipulation and hash map usage. The key insight is that anagrams, when sorted, produce the same string. You need to use this property to group words efficiently. This also tests your ability to work with dictionaries and handle collections of strings.

```python
from collections import defaultdict

class Solution:

    def __init__(self, words):
        self.words = words

    # [["eat", "tea", "ate"], ["tan", "nat"], ["bat"]]
    def solve(self):
        anagrams = defaultdict(set)
        for w in self.words:
            w_sorted = ''.join(sorted(list(w)))
            anagrams[w_sorted].add(w)
        result = []
        for key in anagrams.keys():
            result.append(list(anagrams.get(key)))
        return result

execute = Solution(["eat", "tea", "tan", "ate", "nat", "bat"])
answer = execute.solve()
print(answer)
```
